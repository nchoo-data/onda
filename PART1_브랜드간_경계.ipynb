- 단위가격 수정

```python
import pandas as pd

# 1. CSV 파일 로드
df = pd.read_csv('[0325]통합_제품DF.csv', encoding='cp949')

# 2. [단위가격(100g)] 칼럼 값 삭제 (NaN으로 초기화)
df['단위가격(100g)'] = None

# 3. 단위가격 재계산: 할인가 / 총용량(g) * 100
df['단위가격(100g)'] = df['할인가'] / df['총용량(g)'] * 100

# 방어 코드: 0 나누기 방지
df['단위가격(100g)'] = df.apply(
    lambda row: row['할인가'] / row['총용량(g)'] * 100
    if row['총용량(g)'] and row['총용량(g)'] != 0 else None, axis=1
)
```

```python
df['단위가격(100g)'] = df['단위가격(100g)'].fillna(0).astype(int)
```

# 1. 브랜드별 상품구성 EDA

## 1-1 브랜드별 상품 수

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm


# 브랜드별 상품 수 집계
brand_counts = df['브랜드'].value_counts().reset_index()
brand_counts.columns = ['브랜드', '상품수']

# 색상 설정: '곰곰'은 연하게, 나머지는 진하게
colors = ['#b0c4de' if brand == '곰곰' else '#2c6fad' for brand in brand_counts['브랜드']]

# 시각화
fig, ax = plt.subplots(figsize=(14, 6))

bars = ax.bar(brand_counts['브랜드'], brand_counts['상품수'], color=colors, edgecolor='white', linewidth=0.5)

# 막대 위 수치 표기
for bar, val in zip(bars, brand_counts['상품수']):
    ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.3,
            str(val), ha='center', va='bottom', fontsize=9, color='#333333')

# 범례
from matplotlib.patches import Patch
legend_elements = [
    Patch(facecolor='#2c6fad', label='일반 브랜드'),
    Patch(facecolor='#b0c4de', label='곰곰 (PB 브랜드)')
]
ax.legend(handles=legend_elements, fontsize=10)

ax.set_title('브랜드별 상품 수', fontsize=15, fontweight='bold', pad=15)
ax.set_xlabel('브랜드', fontsize=11)
ax.set_ylabel('상품 수', fontsize=11)
ax.tick_params(axis='x', rotation=45)
ax.set_ylim(0, brand_counts['상품수'].max() * 1.15)
ax.yaxis.grid(True, linestyle='--', alpha=0.5)
ax.set_axisbelow(True)

plt.tight_layout()
plt.show()
```

## 1-2. 브랜드별 분류 분포

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

# 브랜드별 분류 분포 집계
brand_category = df.groupby(['브랜드', '분류']).size().unstack(fill_value=0)
brand_category = brand_category.loc[brand_category.sum(axis=1).sort_values(ascending=False).index]

# 비율 변환
brand_category_pct = brand_category.div(brand_category.sum(axis=1), axis=0) * 100

n_categories = len(brand_category.columns)
base_colors = [cm.tab20(i / n_categories) for i in range(n_categories)]

fig, axes = plt.subplots(1, 2, figsize=(20, 7))

for ax, data, title, ylabel in zip(
    axes,
    [brand_category, brand_category_pct],
    ['브랜드별 분류 분포 (상품 수)', '브랜드별 분류 분포 (비율 %)'],
    ['상품 수', '비율 (%)']
):
    bottom = np.zeros(len(data))

    for i, category in enumerate(data.columns):
        values = data[category].values
        bar_colors = []
        for brand in data.index:
            r, g, b, a = base_colors[i]
            if brand == '곰곰':
                bar_colors.append((r * 0.45 + 0.55, g * 0.45 + 0.55, b * 0.45 + 0.55, a))
            else:
                bar_colors.append((r, g, b, a))
        ax.bar(data.index, values, bottom=bottom, label=category,
               color=bar_colors, edgecolor='white', linewidth=0.5)
        bottom += values

    # 수치 표기
    for i, brand in enumerate(data.index):
        total = data.loc[brand].sum()
        label = f'{total:.1f}%' if ylabel == '비율 (%)' else str(int(total))
        ax.text(i, total + (0.5 if ylabel == '비율 (%)' else 0.2),
                label, ha='center', va='bottom', fontsize=9, color='#333333')

    ax.set_title(title, fontsize=13, fontweight='bold', pad=12)
    ax.set_xlabel('브랜드', fontsize=11)
    ax.set_ylabel(ylabel, fontsize=11)
    ax.tick_params(axis='x', rotation=45)
    ax.set_ylim(0, (105 if ylabel == '비율 (%)' else data.sum(axis=1).max() * 1.15))
    ax.yaxis.grid(True, linestyle='--', alpha=0.4)
    ax.set_axisbelow(True)

# 범례는 우측 차트에만
axes[1].legend(title='분류', bbox_to_anchor=(1.01, 1), loc='upper left', fontsize=10)
axes[0].get_legend_handles_labels()  # 좌측 범례 숨김

plt.suptitle('브랜드별 분류 분포 EDA', fontsize=15, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()
```

## 1-3. 브랜드별 구입가 분포

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

# 제외 브랜드 필터링
exclude_brands = ['차려낸', '99시리즈', '마이퍼스트']
df_eda = df[~df['브랜드'].isin(exclude_brands)].copy()
df_eda = df_eda.rename(columns={'할인가': '구입가'})

# 가격대 구간 설정
bins = [0, 5000, 10000, 20000, 30000, 50000, np.inf]
labels = ['~5천', '5천~1만', '1만~2만', '2만~3만', '3만~5만', '5만+']
df_eda['가격대'] = pd.cut(df_eda['구입가'], bins=bins, labels=labels)

# 브랜드 순서: 총 상품 수 내림차순
brand_order = df_eda['브랜드'].value_counts().index.tolist()
df_eda['브랜드'] = pd.Categorical(df_eda['브랜드'], categories=brand_order, ordered=True)

# 색상 설정
n_labels = len(labels)
base_colors = [cm.tab20(i / n_labels) for i in range(n_labels)]

def get_bar_colors(brands, color):
    r, g, b, a = color
    return [
        (r * 0.45 + 0.55, g * 0.45 + 0.55, b * 0.45 + 0.55, a) if brand == '곰곰' else (r, g, b, a)
        for brand in brands
    ]

fig, axes = plt.subplots(1, 3, figsize=(24, 7))

# ── 차트 1: 브랜드별 가격대 비율 (%) ──────────────────────────────
brand_price = df_eda.groupby(['브랜드', '가격대'], observed=True).size().unstack(fill_value=0)
brand_price = brand_price.loc[brand_order]
brand_price_pct = brand_price.div(brand_price.sum(axis=1), axis=0) * 100

ax1 = axes[0]
bottom = np.zeros(len(brand_price_pct))
for i, label in enumerate(labels):
    if label not in brand_price_pct.columns:
        continue
    values = brand_price_pct[label].values
    bar_colors = get_bar_colors(brand_price_pct.index, base_colors[i])
    ax1.bar(brand_price_pct.index, values, bottom=bottom,
            label=label, color=bar_colors, edgecolor='white', linewidth=0.5)
    bottom += values

ax1.set_title('브랜드별 가격대 비율 (%)', fontsize=12, fontweight='bold', pad=12)
ax1.set_xlabel('브랜드', fontsize=10)
ax1.set_ylabel('비율 (%)', fontsize=10)
ax1.set_ylim(0, 110)
ax1.tick_params(axis='x', rotation=45)
ax1.yaxis.grid(True, linestyle='--', alpha=0.4)
ax1.set_axisbelow(True)
ax1.legend(title='가격대', bbox_to_anchor=(1.01, 1), loc='upper left', fontsize=9)

# ── 차트 2: 브랜드별 구입가 Box Plot ──────────────────────────────
ax2 = axes[1]
box_data = [df_eda[df_eda['브랜드'] == brand]['구입가'].dropna().values for brand in brand_order]
bp = ax2.boxplot(box_data, patch_artist=True, medianprops=dict(color='red', linewidth=1.5),
                 whiskerprops=dict(linewidth=1.2), capprops=dict(linewidth=1.2))

for patch, brand in zip(bp['boxes'], brand_order):
    if brand == '곰곰':
        patch.set_facecolor('#b0c4de')
        patch.set_alpha(0.6)
    else:
        patch.set_facecolor('#2c6fad')
        patch.set_alpha(0.7)

ax2.set_xticks(range(1, len(brand_order) + 1))
ax2.set_xticklabels(brand_order, rotation=45, ha='right', fontsize=9)
ax2.set_title('브랜드별 구입가 Box Plot', fontsize=12, fontweight='bold', pad=12)
ax2.set_xlabel('브랜드', fontsize=10)
ax2.set_ylabel('구입가 (원)', fontsize=10)
ax2.yaxis.grid(True, linestyle='--', alpha=0.4)
ax2.set_axisbelow(True)
ax2.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{int(x):,}'))

# ── 차트 3: 브랜드별 평균가격 vs 중앙값 ──────────────────────────
ax3 = axes[2]
brand_stats = df_eda.groupby('브랜드', observed=True)['구입가'].agg(['mean', 'median']).loc[brand_order]

x = np.arange(len(brand_order))
width = 0.35

mean_colors = ['#b0c4de' if b == '곰곰' else '#2c6fad' for b in brand_order]
med_colors  = ['#d4a5a5' if b == '곰곰' else '#c0392b' for b in brand_order]

bars_mean = ax3.bar(x - width/2, brand_stats['mean'], width, label='평균가격',
                    color=mean_colors, edgecolor='white', linewidth=0.5)
bars_med  = ax3.bar(x + width/2, brand_stats['median'], width, label='중앙값',
                    color=med_colors, edgecolor='white', linewidth=0.5)

ax3.set_xticks(x)
ax3.set_xticklabels(brand_order, rotation=45, ha='right', fontsize=9)
ax3.set_title('브랜드별 구입가 평균 vs 중앙값', fontsize=12, fontweight='bold', pad=12)
ax3.set_xlabel('브랜드', fontsize=10)
ax3.set_ylabel('구입가 (원)', fontsize=10)
ax3.yaxis.grid(True, linestyle='--', alpha=0.4)
ax3.set_axisbelow(True)
ax3.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{int(x):,}'))
ax3.legend(fontsize=10)

plt.suptitle('브랜드별 가격대 분포 EDA', fontsize=15, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()
```

## 2-1. 브랜드 x 분류
- 브랜드 x 분류 독립성 검정 (카이제곱검정)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from scipy.stats import chi2_contingency

# ── 데이터 필터링 ──────────────────────────────────────────────────
exclude_brands = ['곰곰', '차려낸', '99시리즈', '마이퍼스트']
df_2 = df[~df['브랜드'].isin(exclude_brands)].copy()

# ── 카이제곱 검정 ──────────────────────────────────────────────────
contingency_table = pd.crosstab(df_2['브랜드'], df_2['분류'])
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("=" * 55)
print("         브랜드 × 분류 카이제곱 독립성 검정 결과")
print("=" * 55)
print(f"  카이제곱 통계량 : {chi2:.4f}")
print(f"  자유도 (dof)   : {dof}")
print(f"  p-value        : {p:.6f}")
print("-" * 55)
if p < 0.05:
    print("  결론 : p < 0.05 → 브랜드와 분류는 독립적이지 않음")
    print("         (유의수준 5%에서 연관성 존재)")
else:
    print("  결론 : p ≥ 0.05 → 브랜드와 분류는 독립적")
print("=" * 55)

# 크래머의 V (효과 크기)
n = contingency_table.sum().sum()
cramers_v = np.sqrt(chi2 / (n * (min(contingency_table.shape) - 1)))
print(f"  Cramér's V     : {cramers_v:.4f}  ", end="")
if cramers_v < 0.1:
    print("(효과 크기: 매우 작음)")
elif cramers_v < 0.3:
    print("(효과 크기: 작음)")
elif cramers_v < 0.5:
    print("(효과 크기: 중간)")
else:
    print("(효과 크기: 큼)")
print("=" * 55)

# ── 시각화 ─────────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 3, figsize=(24, 7))

# ── 차트 1: 관측 빈도 히트맵 ──────────────────────────────────────
ax1 = axes[0]
obs = contingency_table.values
im1 = ax1.imshow(obs, cmap='Blues', aspect='auto')
ax1.set_xticks(range(len(contingency_table.columns)))
ax1.set_yticks(range(len(contingency_table.index)))
ax1.set_xticklabels(contingency_table.columns, rotation=45, ha='right', fontsize=9)
ax1.set_yticklabels(contingency_table.index, fontsize=9)
for i in range(obs.shape[0]):
    for j in range(obs.shape[1]):
        ax1.text(j, i, str(obs[i, j]), ha='center', va='center',
                 fontsize=9, color='white' if obs[i, j] > obs.max() * 0.6 else '#333333')
plt.colorbar(im1, ax=ax1, shrink=0.8)
ax1.set_title('관측 빈도 (Observed)', fontsize=12, fontweight='bold', pad=12)
ax1.set_xlabel('분류', fontsize=10)
ax1.set_ylabel('브랜드', fontsize=10)

# ── 차트 2: 기대 빈도 히트맵 ──────────────────────────────────────
ax2 = axes[1]
exp = expected
im2 = ax2.imshow(exp, cmap='Oranges', aspect='auto')
ax2.set_xticks(range(len(contingency_table.columns)))
ax2.set_yticks(range(len(contingency_table.index)))
ax2.set_xticklabels(contingency_table.columns, rotation=45, ha='right', fontsize=9)
ax2.set_yticklabels(contingency_table.index, fontsize=9)
for i in range(exp.shape[0]):
    for j in range(exp.shape[1]):
        ax2.text(j, i, f'{exp[i, j]:.1f}', ha='center', va='center',
                 fontsize=9, color='white' if exp[i, j] > exp.max() * 0.6 else '#333333')
plt.colorbar(im2, ax=ax2, shrink=0.8)
ax2.set_title('기대 빈도 (Expected)', fontsize=12, fontweight='bold', pad=12)
ax2.set_xlabel('분류', fontsize=10)
ax2.set_ylabel('브랜드', fontsize=10)

# ── 차트 3: 피어슨 잔차 히트맵 ────────────────────────────────────
ax3 = axes[2]
residuals = (obs - exp) / np.sqrt(exp)
vmax = np.abs(residuals).max()
im3 = ax3.imshow(residuals, cmap='RdBu_r', aspect='auto', vmin=-vmax, vmax=vmax)
ax3.set_xticks(range(len(contingency_table.columns)))
ax3.set_yticks(range(len(contingency_table.index)))
ax3.set_xticklabels(contingency_table.columns, rotation=45, ha='right', fontsize=9)
ax3.set_yticklabels(contingency_table.index, fontsize=9)
for i in range(residuals.shape[0]):
    for j in range(residuals.shape[1]):
        ax3.text(j, i, f'{residuals[i, j]:.2f}', ha='center', va='center',
                 fontsize=9, color='white' if np.abs(residuals[i, j]) > vmax * 0.6 else '#333333')
plt.colorbar(im3, ax=ax3, shrink=0.8)
ax3.set_title('피어슨 잔차 (Pearson Residuals)\n|값| > 2 : 유의미한 편차', fontsize=12, fontweight='bold', pad=12)
ax3.set_xlabel('분류', fontsize=10)
ax3.set_ylabel('브랜드', fontsize=10)

plt.suptitle(f'브랜드 × 분류 카이제곱 독립성 검정  |  χ²={chi2:.2f}, p={p:.4f}, Cramér\'s V={cramers_v:.3f}',
             fontsize=13, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import kruskal, levene

targets_log = ['구입가_log', '총용량(g)_log', '단위가격(100g)_log']
orig_names  = {'구입가_log': '구입가', '총용량(g)_log': '총용량(g)', '단위가격(100g)_log': '단위가격(100g)'}

print("=" * 65)
print("   Kruskal-Wallis — 브랜드별 중앙값 비교 (log 변환)")
print("=" * 65)

for col in targets_log:
    groups = [df_2[df_2['브랜드'] == b][col].dropna().values for b in target_brands]

    # 중앙값 출력
    medians = {b: np.median(g) for b, g in zip(target_brands, groups)}

    # Levene 등분산 검정
    lev_stat, lev_p = levene(*groups)

    # Kruskal-Wallis (중앙값 기반 비모수 검정)
    H, p = kruskal(*groups)

    print(f"\n  ▶ {orig_names[col]} (log 변환)")
    print(f"    중앙값 : { {b: f'{m:.4f}' for b, m in medians.items()} }")
    print(f"    Levene 등분산 검정 : F={lev_stat:.4f}, p={lev_p:.4f}", end="  →  ")
    print("등분산 X" if lev_p < 0.05 else "등분산 O")
    print(f"    Kruskal-Wallis    : H={H:.4f}, p={p:.6f}", end="  →  ")
    if p < 0.001:
        print("*** (p<0.001) 브랜드 간 유의미한 차이 존재")
    elif p < 0.01:
        print("**  (p<0.01)  브랜드 간 유의미한 차이 존재")
    elif p < 0.05:
        print("*   (p<0.05)  브랜드 간 유의미한 차이 존재")
    else:
        print("n.s. 브랜드 간 유의미한 차이 없음")

print("\n" + "=" * 65)
```

### 최적의 클러스터 갯수 선정을 위한 실루엣계수와 산점도 시각화 함수

```python
def visualize_kmeans_plot_multi(cluster_lists, X_features):
    from sklearn.cluster import KMeans
    from sklearn.decomposition import PCA
    import pandas as pd
    import numpy as np
    import matplotlib.pyplot as plt
    import math

    n_clusters = len(cluster_lists)
    n_cols = min(n_clusters, 5)
    n_rows = math.ceil(n_clusters / n_cols)

    fig, axs = plt.subplots(figsize=(4 * n_cols, 4 * n_rows), nrows=n_rows, ncols=n_cols)

    axs = np.array(axs).reshape(-1) if n_clusters > 1 else [axs]

    pca = PCA(n_components=2)
    pca_transformed = pca.fit_transform(X_features)
    dataframe = pd.DataFrame(pca_transformed, columns=['PCA1', 'PCA2'])

    for ind, n_cluster in enumerate(cluster_lists):

        clusterer = KMeans(n_clusters=n_cluster, max_iter=500, random_state=0)
        cluster_labels = clusterer.fit_predict(pca_transformed)
        dataframe['cluster'] = cluster_labels

        unique_labels = np.unique(cluster_labels)
        markers = ['o', 's', '^', 'x', '*', 'D', 'v', 'p', 'H', '+']

        for label in unique_labels:
            label_df = dataframe[dataframe['cluster'] == label]
            cluster_legend = f'Cluster {label}' if label != -1 else 'Noise'
            axs[ind].scatter(x=label_df['PCA1'], y=label_df['PCA2'], s=70,
                             edgecolor='k', marker=markers[label % len(markers)], label=cluster_legend)

        axs[ind].set_title(f'Number of Clusters: {n_cluster}')
        axs[ind].legend(loc='upper right')

    for ax in axs[n_clusters:]:
        ax.axis('off')

    plt.tight_layout()
    plt.show()
```

```python
def visualize_silhouette(cluster_lists, X_features):
    from sklearn.cluster import KMeans
    from sklearn.metrics import silhouette_samples, silhouette_score

    import matplotlib.pyplot as plt
    import matplotlib.cm as cm
    import numpy as np
    import math

    n_clusters = len(cluster_lists)

    n_cols = min(n_clusters, 5)
    n_rows = math.ceil(n_clusters / n_cols)

    fig, axs = plt.subplots(figsize=(4 * n_cols, 4 * n_rows), nrows=n_rows, ncols=n_cols)

    axs = np.array(axs).reshape(-1) if n_clusters > 1 else [axs]

    for ind, n_cluster in enumerate(cluster_lists):

        clusterer = KMeans(n_clusters=n_cluster, max_iter=500, random_state=0)
        cluster_labels = clusterer.fit_predict(X_features)

        sil_avg = silhouette_score(X_features, cluster_labels)
        sil_values = silhouette_samples(X_features, cluster_labels)

        y_lower = 10
        axs[ind].set_title(f'Number of Clusters: {n_cluster}\nSilhouette Score: {round(sil_avg, 3)}')
        axs[ind].set_xlabel("The silhouette coefficient values")
        axs[ind].set_ylabel("Cluster label")
        axs[ind].set_xlim([-0.1, 1])
        axs[ind].set_ylim([0, len(X_features) + (n_cluster + 1) * 10])
        axs[ind].set_yticks([])
        axs[ind].set_xticks([0, 0.2, 0.4, 0.6, 0.8, 1])

        for i in range(n_cluster):
            ith_cluster_sil_values = sil_values[cluster_labels == i]
            ith_cluster_sil_values.sort()

            size_cluster_i = ith_cluster_sil_values.shape[0]
            y_upper = y_lower + size_cluster_i

            color = cm.nipy_spectral(float(i) / n_cluster)
            axs[ind].fill_betweenx(np.arange(y_lower, y_upper), 0, ith_cluster_sil_values,
                                   facecolor=color, edgecolor=color, alpha=0.7)
            axs[ind].text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
            y_lower = y_upper + 10

        axs[ind].axvline(x=sil_avg, color="red", linestyle="--")

    for ax in axs[n_clusters:]:
        ax.axis('off')

    plt.tight_layout()
    plt.show()
```

```python
clus = clustering[['할인율','단위가격(100g)','리뷰수','평균별점','긍정리뷰비율','품질언급비율', '가성비언급비율', '배송언급비율']]
```

```python
visualize_silhouette([3,4,5,6,7], clus)
```
